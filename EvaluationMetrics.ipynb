```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the following lists contain the metrics for all models in order: EfficientNet, CNN, MobileNetV2, VGG16
# Replace the below example values with your actual metrics
accuracy = [90.31, 80.61, 84.21, 68.52]  # Example values
precision = [90.58, 80.61, 84.21, 68.86]  # Example values
recall = [90.31, 80.61, 84.21, 68.53]  # Example values
fscore = [90.30, 80.61, 84.21, 68.38]  # Example values

# Define models for comparison
models = ['EfficientNet', 'CNN', 'MobileNetV2', 'VGG16']

# Prepare DataFrame for visualization
data = [
    ['EfficientNet', 'Precision', precision[0]], ['EfficientNet', 'Recall', recall[0]], ['EfficientNet', 'F1 Score', fscore[0]], ['EfficientNet', 'Accuracy', accuracy[0]],
    ['CNN', 'Precision', precision[1]], ['CNN', 'Recall', recall[1]], ['CNN', 'F1 Score', fscore[1]], ['CNN', 'Accuracy', accuracy[1]],
    ['MobileNetV2', 'Precision', precision[2]], ['MobileNetV2', 'Recall', recall[2]], ['MobileNetV2', 'F1 Score', fscore[2]], ['MobileNetV2', 'Accuracy', accuracy[2]],
    ['VGG16', 'Precision', precision[3]], ['VGG16', 'Recall', recall[3]], ['VGG16', 'F1 Score', fscore[3]], ['VGG16', 'Accuracy', accuracy[3]],
]

df = pd.DataFrame(data, columns=['Algorithms', 'Parameters', 'Value'])

# Pivot DataFrame for plotting
pivot_df = df.pivot(index='Parameters', columns='Algorithms', values='Value')

# Plot bar chart
pivot_df.plot(kind='bar', figsize=(10, 6))
plt.title("Comparison of EfficientNet, CNN, MobileNetV2, and VGG16 Performance")
plt.ylabel("Metrics Value (%)")
plt.xlabel("Parameters")
plt.legend(title="Models")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Performance of all algorithms in tabular format
columns = ["Algorithm Name", "Accuracy", "Precision", "Recall", "F1-Score"]
values = []

# Algorithm names
algorithm_names = ['EfficientNet', 'CNN', 'MobileNetV2', 'VGG16']

# Append metrics for each algorithm
for i in range(len(algorithm_names)):
    values.append([algorithm_names[i], accuracy[i], precision[i], recall[i], fscore[i]])

# Create a DataFrame
performance_table = pd.DataFrame(values, columns=columns)

# Display the table
performance_table
```
